```python
import numpy as np
import matplotlib.pyplot as plt
import codecs
#size=25
size=25
params = {'legend.fontsize': size*.75,
          'axes.labelsize': size,
          'axes.titlesize': size,
          'xtick.labelsize': size*0.75,
          'ytick.labelsize': size*0.75,
          'axes.titlepad': 25}
plt.rcParams.update(params)

xxs = []
yys = []
for suffix in ['2', '4', '6', '8']:
    xs = []
    ys = []
    gx = codecs.open(f'gX{suffix}.txt')
    pi = codecs.open(f'pi{suffix}.txt')
    pi2 = codecs.open(f'pi2{suffix}.txt')
    ramsey = codecs.open(f'ramsey{suffix}.txt')
    for file in [gx, pi, pi2, ramsey]:
        file_data = np.loadtxt(file, usecols=(0,1),skiprows=1)
        x = file_data[:,0]
        y = file_data[:,1]
        xs.append(x)
        ys.append(y)
    xxs.append(xs)
    yys.append(ys)
```

```python
cs = ['k', 'g', 'y', 'r']
for j, xx in enumerate(xxs):
    for i, x in enumerate(xx):
        plt.plot(x, yys[j][i], c=cs[i], label=['gX', '', '/2', 'Ramsey'][i])
    plt.xlabel("")
    plt.ylabel("J()")
    plt.ylim(0, 70)
    plt.xlim(0, 7)
    plt.legend()
    plt.show()
```
